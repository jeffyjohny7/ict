```python
def checkOutlier(a, b):
    Q1 = a[b].quantile(0.25)
    Q3 = a[b].quantile(0.75)

    IQR = Q3 - Q1

    width = 1.5

    checkOutlier.Max = Q3 + (width * IQR)
    checkOutlier.Min = Q1 - (width * IQR)

    print("Maximum Value:", checkOutlier.Max)
    print("Minimum Value:", checkOutlier.Min)
```

```python
for col in ['ball_controlling_skills','jumping_skills', 'penalties_conversion_rate', 'mental_strength','shot_accuracy', 'behaviour_rating',
       'matches_played', 'fitness_rating','years_of_experience']:
    result = checkOutlier(data, col)
    print(f"Outliers of column {col} will be not there in the range of minimum and maximum value")
```

```python
def removeOutliers(a, b):
    checkOutlier(a, b)


    outliers_removed = a[(a[b] >= checkOutlier.Min) & (a[b] <= checkOutlier.Max)]

    return outliers_removed


columns_to_check = ['ball_controlling_skills', 'jumping_skills', 'penalties_conversion_rate', 'mental_strength',
                    'shot_accuracy', 'behaviour_rating', 'matches_played', 'fitness_rating', 'years_of_experience']

for col in columns_to_check:
    data = removeOutliers(data, col)
```
